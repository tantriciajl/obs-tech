```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# PART A
ds = pd.read_csv('galaxies.csv')

ug_a = ds['u'] - ds['g']
gr_a = ds['g'] - ds['r']

plt.plot(ug_a, gr_a, '.')
plt.xlabel('u - g', fontsize=14)
plt.ylabel('g - r', fontsize=14)
plt.title('Part A: 9 Galaxies', fontsize=16)

# plot y = -x + 2.2
x = np.linspace(min(ug_a), max(ug_a), 100)
y = -x + 2.2

plt.plot(x, y, label='y = -x + 2.2')
plt.legend(loc='best', fontsize=14)

# PART B
df = pd.read_csv('partb.csv')

ug_b = df['u'] - df['g']
gr_b = df['g'] - df['r']

plt.figure()
plt.plot(ug_b, gr_b, '.', label='Chosen Galaxies')
plt.xlabel('u - g', fontsize=14)
plt.ylabel('g - r', fontsize=14)
plt.title('Part B: 20 Galaxies', fontsize=16)

type3 = np.loadtxt('type3.csv', skiprows=2, delimiter=',')

type3[:,6]

ug_radial = type3[:,9] - type3[:,10]
gr_radial = type3[:,10] - type3[:,11]

plt.plot(ug_radial, gr_radial, 'r.', label='Radial Search Galaxies')

plt.plot(x, y, label='y = -x + 2.2')

plt.legend(loc='best')
```
